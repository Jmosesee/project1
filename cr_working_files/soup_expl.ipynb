```python
from bs4 import BeautifulSoup

import requests

url = 'https://www.indeed.com/jobs?q=data%20scientist&l=Denver%2C%20CO&vjk=da4412e0f36920a0'

r  = requests.get(url)

data = r.text

soup = BeautifulSoup(data)

for link in soup.find_all('a'):
    print(link.get('href'))
```
